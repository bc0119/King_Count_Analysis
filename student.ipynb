```python
def binary_average(row):
    
    if row == '7 Average':
        return 1
    else:
        return 0
    
def binary_low_average(row):
    
    if row == '6 Low Average':
        return 1
    else:
        return 0
    
def binary_good(row):
    
    if row == '8 Good':
        return 1
    else:
        return 0

def binary_excellent(row): 
    if row == '11 Excellent':
        return 1
    else:
        return 0
    
def binary_better(row):
    if row == '9 Better':
        return 1
    else:
        return 0
    
def binary_fair(row): 
    if row == '5 Fair':
        return 1
    else:
        return 0
    
def binary_very_good(row):
    
    if row == '10 Very Good':
        return 1
    else:
        return 0
    
def binary_luxury(row):
    
    if row == '12 Luxury':
        return 1
    else:
        return 0
    
def binary_low(row):
    
    if row == '4 Low':
        return 1
    else:
        return 0
    
def binary_poor(row):
    
    if row == '3 Poor':
        return 1
    else:
        return 0
    
def binary_mansion(row):
    
    if row == '13 Mansion':
        return 1
    else:
        return 0
```

```python
data['Average'] = data.grade.apply(lambda x: binary_average(x))
data['Low_Average'] = data.grade.apply(lambda x: binary_low_average(x))
data['Good'] = data.grade.apply(lambda x: binary_good(x))
data['Excellent'] = data.grade.apply(lambda x: binary_excellent(x))
data['Better'] = data.grade.apply(lambda x: binary_better(x))
data['Fair'] = data.grade.apply(lambda x: binary_fair(x))
data['Very_Good'] = data.grade.apply(lambda x: binary_very_good(x))
data['Luxury'] = data.grade.apply(lambda x: binary_luxury(x))
data['Low'] = data.grade.apply(lambda x: binary_low(x))
data['Poor'] = data.grade.apply(lambda x: binary_poor(x))
data['Mansion'] = data.grade.apply(lambda x: binary_mansion(x))
```

```python
def condition_ranking(row):
    
    if row == 'Average':
        return 3
    elif row == 'Good':
        return 4
    elif row == 'Very Good':
        return 5
    elif row == 'Fair':
        return 2
    else:
        return 1

data['condition_ranking']= data.condition.apply(lambda x: condition_ranking(x))
plt.scatter(data.price, data.condition_ranking)
plt.show()
```

```python
data.view.unique()
```

```python
def view_ranking(row):
    if row == 'NONE':
        return 0
    elif row == 'GOOD':
        return 3
    elif row == 'EXCELLENT':
        return 5
    elif row == 'AVERAGE':
        return 2
    else:
        return 1
    
data['view_ranking'] = data.view.apply(lambda x: view_ranking(x))
data.view_ranking.value_counts()
```

```python
data.columns
```

```python
def renovated_binary(row):
    if row != 0:
        return 1
    else:
        return 0
    
data['renovated_binary'] = data.yr_renovated.apply(lambda x: renovated_binary(x))
```

```python
def grade_ranking(row):
    if row == '3 Poor':
        return 1
    elif row == '4 Low':
        return 2
    elif row == '5 Fair':
        return 3
    elif row == '6 Low Average':
        return 4
    elif row == '7 Average':
        return 5
    elif row == '8 Good':
        return 6
    elif row == '9 Better':
        return 7
    elif row == '10 Very Good':
        return 8
    elif row == '11 Excellent':
        return 9
    elif row == '12 Luxury':
        return 10
    else:
        return 11
    

data['grade_ranking'] = data.grade.apply(lambda x: grade_ranking(x))
data.grade_ranking.value_counts()
```
